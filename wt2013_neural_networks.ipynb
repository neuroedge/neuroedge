The four books are:  

1. **Introduction to the Theory of Neural Computation** by Hertz. Prof Simen showed this to me and said it might be useful and that I could code an example from there.
The book is geared towards statistical physicists, and so was not useful to people without similar mathematical background.  

2. **Neural Networks for Pattern Recognition** by Bishop. Hinton (who made a Coursera course on Neural Networks & Machine Learning) wrote a foreword for this book, 
remarking on its ease of use. Later, a friend who graduated from cambridge wondered why I was reading this 'hard-core' AI book. I was not sure myself.

3. **Theoretical Neuroscience**,  by Dayan. This is one of the classic texts, and is in use by MIT for its graduate class on introductory theoretical neuroscience.
It covers modelling and abstraction from the neuron level to the network level. It also covers neural encoding and decoding, and provides a chapter on Shannon's Information Theory.
This text comes with problems, which are available online at http://www.gatsby.ucl.ac.uk/~dayan/book/exercises.html  

4. **Principles of Computational Modelling in Neuroscience** by Sterratt. The book which I feel is most likely something that can be coverred at an undergraduate level.
It is well-designed and every sentence follows each other, and the equation does not separate themselves from the text. It is easy to follow. 
It covers the same ground as **Mathematics for Neuroscientists** by Gabbiani. Likewise, it places much less emphasis on networks, more on individual neurons, starting from membranes. 
But unlike Gabbiani, this is a better book, and at 11 chapters and 300+ pages, is an easier read than Gabbiani.

```python
from IPython.core.display import display
 
class YouTubeVideo(object):
    """Class for embedding a YouTube Video in an IPython session, based on its video id.
    Credits to pastebin.com/W6xsNgZD
    e.g. to embed the video on this page:
   
    http://www.youtube.com/watch?v=foo
   
    you would do:
   
    vid = YouTubeVideo("foo")
    display(vid)
    """
   
    def __init__(self, id, width=400, height=300):
        self.id = id
        self.width = width
        self.height = height
   
    def _repr_html_(self):
        """return YouTube embed iframe for this video id"""
        return """
            <iframe
                width="%i"
                height="%i"
                src="http://www.youtube.com/embed/%s"
                frameborder="0"
                allowfullscreen
            ></iframe>
        """%(self.width, self.height, self.id)
```
